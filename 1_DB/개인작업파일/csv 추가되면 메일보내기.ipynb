```python
# 키워드가 df_total의 TITLE_NM컬럼과 겹치는 경우

import re

for title in df_total["TITLE_NM"]:
    for keyword in df_test["SRCHWRD"]:
        if re.search(keyword, title, re.IGNORECASE):
            print(f"비슷한 글자 발견 - title: {title}, keyword: {keyword}")
```

```python
import re
filtered_rows = []

# df_test["SRCHWRD"]가 주어진 df_total의 컬럼들("TITLE_NM", "AUTHR_NM", "PUBLISHER_NM", "BOOK_INTRCN_CN", "Combined", "KDC_1")에서 비슷한 글자가 있는지 확인
for index, row in df_total.iterrows():
    for column in ["TITLE_NM", "AUTHR_NM", "PUBLISHER_NM", "BOOK_INTRCN_CN", "Combined","KDC_1"]:
        for keyword in df_test["SRCHWRD"]:
            if re.search(keyword, str(row[column]), re.IGNORECASE):
                filtered_rows.append(row)
                break

# 필터를 거쳐서 새로운 데이터프레임 생성
df_filtered = pd.DataFrame(filtered_rows)

# 결과 확인
print(df_filtered)
```

```python
import re
filtered_rows = []

df_total = df_total.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

for index, row in df_total.iterrows():
    for column in ["TITLE_NM", "AUTHR_NM", "PUBLISHER_NM", "BOOK_INTRCN_CN", "Combined","KDC_1"]:
        for index2, row2 in df_test.iterrows():
           if row["KDC_1"] == row2["KDC"]:
              for keyword in df_test.loc[index2, "SRCHWRD"]:
                  if re.search(keyword, str(row[column]), re.IGNORECASE):
                     filtered_rows.append(row)
                     break

# 필터를 거쳐서 새로운 데이터프레임 생성
df_filtered = pd.DataFrame(filtered_rows)

# 결과 확인
print(df_filtered)
```

```python
import pandas as pd

# df_total과 df_test를 merge하여 필요한 컬럼만 남기기
df_merged = df_total.merge(df_test, left_on="KDC_1", right_on="KDC")

# 필터링을 위한 정규식 패턴 생성
pattern = "|".join(df_merged["SRCHWRD"].str.replace(" ", r"\s*").tolist())

# 필터링 수행
mask = df_merged[["TITLE_NM", "AUTHR_NM", "PUBLISHER_NM", "BOOK_INTRCN_CN", "Combined"]].apply(
    lambda x: x.str.contains(pattern, case=False, regex=True)
).any(axis=1)

# 결과 데이터프레임 생성
df_filtered = df_merged[mask]

# 필요한 컬럼만 남기기
df_filtered = df_filtered[["TITLE_NM", "AUTHR_NM", "PUBLISHER_NM", "BOOK_INTRCN_CN", "Combined", "KDC_1"]]

# 결과 확인
print(df_filtered)
```

```python
import random

# 데이터프레임에서 "KDC_1" 번호가 같은 행을 무작위로 선정하는 함수
def select_random_rows(group):
    if len(group) >= 3:
        return group.sample(3)
    else:
        return group

# "KDC_1" 번호가 같은 행을 무작위로 선정하여 새로운 데이터프레임 생성
df_random_sample = df_filtered.groupby("KDC_1", group_keys=False).apply(select_random_rows)

# 결과 확인
print(df_random_sample)
```
