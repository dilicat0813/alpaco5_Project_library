```python
df1 = pd.read_csv("data/도서 키워드/NL_BO_TAG_CLOUD4_202012-1.csv")
df2 = pd.read_csv("data/도서 키워드/NL_BO_TAG_CLOUD4_202012-2.csv")
df3 = pd.read_csv("data/도서 키워드/NL_BO_TAG_CLOUD4_202012-3.csv")
df4 = pd.read_csv("data/도서 키워드/NL_BO_TAG_CLOUD4_202012-4.csv")
df5 = pd.read_csv("data/도서 키워드/NL_BO_TAG_CLOUD4_202012-5.csv")
df6 = pd.read_csv("data/도서 키워드/NL_BO_TAG_CLOUD4_202012-6.csv")

df = pd.concat([df1,df2,df3,df4,df5,df6])
```

```python
print(len(df))
```

```python
df
##FQ_CO를 캇 하는 기준을.. 중간/평균값 중 하나로 잡아야 되고
# 이 수많은 데이터들 중 중간값으로 캇 할수있는지??
#pivot_table = df.pivot_table(index='날짜', columns='지역', values='상품', aggfunc=combine_strings)
# 피봇 테이블 생성
def combine_strings(series):
    return ', '.join(series)

pvdf = df.pivot_table(index='ISBN_NO', columns='FQ_CO', values='WRD_NM', aggfunc=combine_strings)
```

```python
flat_table = pvdf.unstack().reset_index()
flat_table.columns = ['FQ_CO', 'ISBN_NO', 'Combined']
flat_table.dropna(inplace=True)
# 결과 출력
```

```python
flat_table
```

```python
def combine_strings(group):
    return ', '.join(group)

new_df = flat_table.groupby('ISBN_NO')['Combined'].agg(combine_strings).reset_index()
```

```python
new_df.to_csv("searchword.csv")
```
