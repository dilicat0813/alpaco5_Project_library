```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('data/NL_LON_HALFLIFE_2020.csv')
df2 = pd.read_csv('data/NL_ON_SPECIES_ANALYSIS_MASTER_202112.csv')
```

```python
df1['BOOK_MASTR_SEQ_NO'] =df1['BOOK_MASTR_SEQ_NO'].astype(float)
# df1['ISBN_NO'] = df1['ISBN_NO'].astype(str)
df1.info()
```

```python
df2['BOOK_MASTR_SEQ_NO'] = df2['BOOK_MASTR_SEQ_NO'].astype(float)
# df2['ISBN_THIRTEEN_NO'] = df2['ISBN_THIRTEEN_NO'].astype(str)
df2.info()
```

```python
# 두 데이터프레임을 머지
merged_df = pd.merge(df1, df2, left_on='BOOK_MASTR_SEQ_NO', right_on='BOOK_MASTR_SEQ_NO', how='inner')
# merged_df = pd.merge(df1, df2, left_on='ISBN_NO', right_on='ISBN_THIRTEEN_NO', how='inner')
# merged_df['BOOK_CO'].unique()
merged_df
```
